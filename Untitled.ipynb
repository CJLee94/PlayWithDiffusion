# Forward Process of Denoising Diffusion Model

$q(z_{\lambda}|x) = \mathcal{N}(\alpha_{\lambda}\mathbf{x}, \sigma_{\lambda}^2\mathbf{I})$ where $\alpha_{\lambda}^2 = \frac{1}{1+e^{-\lambda}}, \sigma_{\lambda}^2=1-\alpha^2_{\lambda}$

```python
import numpy as np
import matplotlib.pyplot as plt

def forward(x, lam):
    # forward process to add noise to the input x at step lam
    alpha_square = 1/(1+np.exp(-lam))
    sigma_square = 1-alpha_square

    # return np.sqrt(alpha_square)*x + sigma_square*np.random.randn(*x.shape)
    return np.sqrt(alpha_square)*x + np.sqrt(sigma_square)*np.random.randn(*x.shape)

# def forward_sbs(x, lam_p, lam):
    
```

Let us generate samples from a gaussian mixture distribution

```python
N_sample = 2000 # number of samples

N_mixture = 3 # number of mixutres

W = np.random.rand(N_mixture) # weight of each mixture
W = W/W.sum()

mu = 6*(np.random.rand(N_mixture)-0.5)
sigma = np.random.rand(N_mixture)/6

# def generate_samples():
sample_mixture = np.random.choice(3, size=N_sample, p=W)
sample_mu = mu[sample_mixture]
sample_sigma = sigma[sample_mixture]

samples = sample_mu+sample_sigma*np.random.randn(N_sample)
lower = samples.min()
upper = samples.max()
max_range = max(abs(lower), abs(upper))
lower = -5
upper = 5

print(f"The sampled mean of each mixture {mu}")
print(f"The sampled sigma of each mixture {sigma}")

_ = plt.hist(samples, bins = N_sample//8)
```

```python
nbins = 100
N_step = 1000

lam_max = 20
lam_min = -10
b = np.arctan(np.exp(-lam_max/2))
a = np.arctan(np.exp(-lam_min/2))-b
u_range = np.linspace(0,1,1000)
lam_range = -2*np.log(np.tan(a*u_range+b))

sample_mat = np.zeros((nbins, N_step+1))
sample_hist = np.histogram(samples, bins = nbins, range=(lower, upper))[0]
sample_mat[:,0] = sample_hist/sample_hist.max()
normal_mat = np.zeros((nbins, N_step+1))
normal_hist = np.histogram(np.random.randn(len(samples)), bins=nbins, range=(lower, upper))[0]
normal_mat[:,0] = normal_hist/normal_hist.max()

for idx, lam in enumerate(lam_range):
    lam_samples = forward(samples, lam)
    sample_hist = np.histogram(lam_samples, bins = nbins, range=(lower, upper))[0]
    sample_mat[:,idx+1] = sample_hist/sample_hist.max()
    normal_hist = np.histogram(np.random.randn(len(samples)), bins=nbins, range=(lower, upper))[0]
    normal_mat[:,idx+1] = normal_hist/normal_hist.max()
plt.figure()
plt.imshow(sample_mat)
plt.figure()
plt.imshow(normal_mat)
```
